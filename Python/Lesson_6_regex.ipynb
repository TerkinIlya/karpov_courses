```python
import re
```

```python
banks = '''бета банк
бетабанк
Бета-банк
БеттаБанк
бэтта банк
Сбербанк
бета онлайн банка
бета  банк'''
```

```python
bank_pattern = re.compile('б[еэ]т{1,2}а.{0,2}банк')
```

```python
bank_pattern.findall(banks)
```

```python
bank_pattern = re.compile('б[еэ]т{1,2}а.{0,2}банк', re.IGNORECASE | re.MULTILINE)
```

```python
bank_pattern.findall(banks)
```

```python
text = 'Этот текст содержит знаки препинания, а так же цифры 1 и 0. Цель занятия: разбор вопросов.'
```

```python
text_pattern = re.compile(r'[^\w\s]')
```

```python
text_pattern.findall(text)
```

```python
re.sub(text_pattern, '', text)
```
