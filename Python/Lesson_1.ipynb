```python
workers = [['Ivan', 'Ivanov', 100000, 1], ['Petr', 'Petrov', 150000, 2], ['Sidor', 'Sidorov', 200000, 10]]
```

```python
def check_status(exp):
    if exp > 5:
        position = 'senior'
    elif exp < 2:
        position = 'junior'
    else:
        position = 'middle'
    return position

for worker in workers:
    print(f'{worker[0]} {worker[1]} is {check_status(worker[-1])}')
```
