```python
import pandas as pd
```

```python
all_data = []
```

```python
with open('/content/drive/MyDrive/Karpov_Courses/data/meals_list_2.txt', 'r', encoding='utf-8') as f:
    scaning = False
    for line in f.readlines():
      if 'Non-deleted saved meals' in line:
        scaning = True
        continue
      if line.startswith('-------------'):
        scaning = False
      if scaning:
        if 'Meal#' in line:
          col_name = line.split(';')
          continue
        all_data.append([value.strip() for value in line.split(';')])
```

```python
col_name = [col.strip() for col in col_name]
```

```python
col_name
```

```python
all_data
```

```python
meal_df = pd.DataFrame(all_data, columns=col_name[:-1])
```
