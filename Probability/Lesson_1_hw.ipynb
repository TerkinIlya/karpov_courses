### Task 1
Два человека одновременно стреляют по мишени. Вероятность попадания по мишени у первого стрелка равна 0,5, у второго — 0,8. Какова вероятность того, что в мишени будет только одна пробоина?

```python
prob_1_hit = 0.5 * (1 - 0.8)
prob_2_hit = 0.8 * (1 - 0.5)
prob_1_hit + prob_2_hit
```

### Task 2
Трое аналитиков на экзамене независимо друг от друга решают одну и ту же задачу. Вероятности ее решения этими аналитиками равны 0,8, 0,7 и 0,6 соответственно. Найдите вероятность того, что хотя бы один аналитик решит задачу.

```python
prob_not_solve = (1 - 0.8) * (1 - 0.7) * (1 - 0.6)
print(prob_not_solve)
1 - prob_not_solve
```

### Task 3

### Task 5
Из колоды в 36 карт наудачу извлекается одна карта. События:

A = {вынутая карта – туз}
В = {вынутая карта чёрной масти}
F = {вынутая карта – фигура, т.е. является валетом, дамой, королём или тузом}.

Выберите все пары зависимых событий:


A и F

B и F

А и В


События А и F зависимы, так как все тузы являются фигурами, поэтому если вынут туз, то это также будет событием F.

События А и В тоже независимы, так как туз может быть как чёрным, так и красным, поэтому наличие туза не влияет на цвет вынутой карты.

Таким образом, зависимыми являются только события A и F.

### Task 7

В трех углах равностороннего треугольника находится по роботу. Каждый из роботов начинает двигаться в случайно выбранный угол по стороне треугольника. Какова вероятность того, что ни один из роботов не столкнется с другим роботом?

Ответ округлите до сотых. В качестве десятичного разделителя используйте точку.

Рассмотрим движение каждого робота. Он может двигаться только в одном из двух направлений, т.е. в каждый момент времени каждый робот делает выбор из двух направлений движения.

Всего возможно 2 * 2 * 2 = 8 вариантов направлений для трех роботов.

Чтобы никакие два робота не столкнулись, каждый робот должен двигаться в направлении, противоположном направлению движения одного из других роботов. Таким образом, для того, чтобы ни один из роботов не столкнулся, каждый из них должен выбрать одно из двух направлений, не выбранных другими роботами.

Первый робот может выбрать направление движения любым из двух способов. Второй робот может выбрать направление движения только одним из двух оставшихся направлений. Аналогично, третий робот может выбрать направление движения только одним из двух оставшихся направлений.

Таким образом, всего существует 2 * 1 * 1 = 2 вариантов, когда все три робота движутся в направлениях, не пересекающихся.

Вероятность того, что никакие два робота не столкнутся, равна 2/8 или 1/4.

Ответ: вероятность того, что ни один из роботов не столкнется с другим роботом, равна 1/4 или 0,25.
