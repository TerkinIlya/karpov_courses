## Задание 1

Из 500 компьютеров  180 принадлежат к 1 партии, 170 – ко второй партии, остальные к третьей. В первой партии 3% брака, во второй — 2%, в третьей – 6%. Случайно выбирается один компьютер. Определить вероятность того, что выбранный компьютер – бракованный.  Округлите полученный ответ до десятитысячных. В качестве разделителя используйте точку.

```python
def full_prob(quantity_of_items: tuple, probs: tuple, rounding: int = None, proportion: bool = False) -> float:
  result = 0
  if proportion:
    amount_items = 1
  else:
    amount_items = sum(quantity_of_items)
  for it, prob in zip(quantity_of_items, probs):
    result += it / amount_items * prob
  if rounding:
     result = round(result, rounding)
  return result
```

```python
computers = (180, 170, 500 - (180 + 170))
probs = (0.03, 0.02, 0.06)
full_prob(computers, probs, 4)
```

## Задание 3

Из 40 снайперов 18 попадает в цель с вероятностью 0,9, 8 — с вероятностью 0,4 и 14 – с вероятностью 0,7. Случайно выбранный снайпер произвел выстрел, поразив цель. Мы хотим вычислить, снайпер из какой группы наиболее вероятно совершил этот выстрел.

В ответ запишите наибольшую вероятность, округленную до тысячных. В качестве разделителя используйте точку.

```python
def bayes_theorem(quantity_of_items: tuple, probs: tuple, group: int, rounding: int = None, proportion: bool = False) -> float:
  index = group - 1
  prob_of_group = quantity_of_items[index] / sum(quantity_of_items)
  result = (prob_of_group * probs[index]) / full_prob(quantity_of_items, probs, proportion=proportion)
  if rounding:
    result = round(result, rounding)
  return result
```

```python
sniper = (18, 8, 14)
probs = (0.9, 0.4, 0.7)
for i in range(1, 4):
  print(bayes_theorem(sniper, probs, i, 3))
```

```python
from scipy import special

def bernulli(n: int, k: int, p: int|float) -> float:
  q = 1 - p
  return special.comb(n, k) * p ** k * q ** (n - k)
```

```python
group_1_prob = bernulli(7, 1, 1/2)
group_2_prob = bernulli(7, 1, 1/6)
```

## Задание 6

Известно, что около 60% всех людей в городе болеют заболеванием А. Для определения того, заболел человек или нет, проводится тест, точность которого составляет 99%. Вы получили положительный тест, какова вероятность того, что вы реально больны?

Округлите ответ до тысячных, в качестве десятичного раделителя используйте точку.

Для решения этой задачи мы можем использовать теорему Байеса. Давайте обозначим следующие события:

A - человек действительно болен заболеванием А
B - положительный результат теста

Мы ищем вероятность того, что человек действительно болен при условии положительного теста, то есть P(A|B).

Известно, что 60% всех людей в городе болеют заболеванием А, так что P(A) = 0,6. Точность теста составляет 99%, что означает, что вероятность ложноположительного результата (человека, который не болен, но получает положительный тест) составляет 1% или 0,01. Следовательно, вероятность получить положительный тест при условии, что вы действительно болеете, равна точности теста, то есть P(B|A) = 0,99.

С использованием теоремы Байеса мы можем выразить P(A|B) следующим образом:

P(A|B) = (P(B|A) * P(A)) / P(B)

Теперь давайте найдем P(B), вероятность положительного теста. Эта вероятность может быть вычислена с использованием полной вероятности и формулы:

P(B) = P(B|A) * P(A) + P(B|¬A) * P(¬A)

P(B|¬A) - это вероятность положительного результата теста при условии, что человек не болен, а P(¬A) - вероятность быть здоровым. В данном случае P(¬A) = 1 - P(A) = 1 - 0,6 = 0,4. Поскольку точность теста равна 99%, P(B|¬A) = 1 - точность теста = 1 - 0,99 = 0,01.

Теперь мы можем вычислить P(B):

P(B) = P(B|A) * P(A) + P(B|¬A) * P(¬A)
= 0,99 * 0,6 + 0,01 * 0,4
= 0,594 + 0,004
= 0,598

Теперь мы можем выразить P(A|B):

P(A|B) = (P(B|A) * P(A)) / P(B)
= (0,99 * 0,6) / 0,598
≈ 0,98996655518

Таким образом, при положительном тесте вероятность того, что вы реально больны, составляет примерно 0,98997 или около 98,997%.


```python
test_accuracy = 0.99
sick = 0.60
healthy = 0.40

prob_test_positive = sick * test_accuracy + healthy * (1 - test_accuracy)
#prob_test_positive
round((sick * test_accuracy) / prob_test_positive, 3)
```

## Задание 8

Система РЛС (радиолокационная система) может обнаружить присутствие самолета в 98% случаев. Однако если в зоне действия радара нет самолета, система все равно сообщает (ложно), что самолет присутствует, в 5% случаев. В любой момент времени вероятность того, что самолет находится в зоне действия радара, составляет 7%.Какова вероятность того, что самолет находится в зоне действия радара, учитывая, что на радаре нет самолета?

Ответ округлите до десятитысячных, в качестве десятичного разделителя используйте точку.

Да сам удивляюсь, не иначе звёзды сошлись.

- $P(Y|X)$ - вероятность, что самолёт есть, если радар его не видит (**то, что мы ищем в задаче**)
- $P(X|Y)$ - вероятность, что радар ничего не увидит, если самолёт прилетит (**в условии указана обратная вероятность - что радар увидит реально летящий самолёт**)
- $P(Y)$ - вероятность, что самолёт в принципе может залететь в область радара (**также указана в условии**)
- $P(X)$ - вероятность того, что самолёт не будет видно на радаре в принципе (**считаем по аналогии с прошлыми задачами, учитываем указанную в условии вероятность ложноположительного срабатывания, но про ложноотрицательное тоже не забывайте**)

Давайте введем следующие обозначения:
A - самолет находится в зоне действия радара,
B - радар сообщает о наличии самолета.

Нам известны следующие данные:
P(A) = 0.07 - вероятность того, что самолет находится в зоне действия радара,
P(B|A) = 0.98 - вероятность того, что радар обнаружит самолет, если он действительно присутствует,
P(B|¬A) = 0.05 - вероятность того, что радар сообщит о наличии самолета, если его на самом деле нет.

Нам требуется найти вероятность P(A|¬B), то есть вероятность того, что самолет находится в зоне действия радара, учитывая, что радар не сообщил о его присутствии.

Мы можем использовать формулу условной вероятности для этого:

P(A|¬B) = (P(¬B|A) * P(A)) / P(¬B),

где P(¬B|A) - вероятность того, что радар не сообщит о наличии самолета, если он действительно присутствует,
P(¬B) - вероятность того, что радар не сообщит о наличии самолета.

Чтобы вычислить P(¬B|A), мы можем использовать дополнение:

P(¬B|A) = 1 - P(B|A) = 1 - 0.98 = 0.02.

Теперь давайте вычислим P(¬B) с помощью формулы полной вероятности:

P(¬B) = P(¬B|A) * P(A) + P(¬B|¬A) * P(¬A),

где P(¬B|¬A) - вероятность того, что радар не сообщит о наличии самолета, если его действительно нет,
P(¬A) - вероятность того, что самолет не находится в зоне действия радара.

P(¬B|¬A) = 1 - P(B|¬A) = 1 - 0.05 = 0.95,
P(¬A) = 1 - P(A) = 1 - 0.07 = 0.93.

Теперь мы можем вычислить P(¬B):

P(¬B) = P(¬B|A) * P(A) + P(¬B|¬A) * P(¬A)
= 0.02 * 0.07 + 0.95 * 0.93
= 0.0014 + 0.8835
= 0.8849.

Теперь мы можем использовать найденные значения для вычисления P(A|¬B):

P(A|¬B) = (P(¬B|A) * P(A)) / P(¬B)
= (0.02 * 0.07) / 0.8849
= 0.0014 / 0.8849
≈ 0.0016.

Таким образом, вероятность того, что самолет находится в зоне действия радара, учитывая, что на радаре нет самолета, составляет около 0.0016 или около 0.16%.

```python
P_XY = 1 - 0.98
P_Y = 0.07
false_pos = 0.05
P_X = P_XY * P_Y + (1 - false_pos) * (1 - P_Y)
result = (P_XY * P_Y) / P_X
round(result, 4)
```

```python
items_prob = (0.07, 1 - 0.07)
probs = (1 - 0.98, 1 - 0.05)
bayes_theorem(items_prob, probs, 1, 4, proportion=True)
```

## Задание 9

Система РЛС (радиолокационная система) может обнаружить присутствие самолета в 98% случаев. Однако если в зоне действия радара нет самолета, система все равно сообщает (ложно), что самолет присутствует, в 5% случаев. В любой момент времени вероятность того, что самолет находится в зоне действия радара, составляет 7%.Какова вероятность того, что самолет находится в зоне действия радара, учитывая, что радар показывает наличие самолета?

Ответ округлите до тысячных, в качестве десятичного разделителя используйте точку.

Для решения этой задачи воспользуемся формулой Байеса:

P(A|B) = (P(B|A) * P(A)) / P(B),

где
P(A|B) - вероятность того, что самолет находится в зоне действия радара, учитывая, что радар показывает наличие самолета,
P(B|A) - вероятность того, что радар показывает наличие самолета, если самолет действительно находится в зоне действия радара,
P(A) - вероятность того, что самолет находится в зоне действия радара,
P(B) - вероятность того, что радар показывает наличие самолета.

Из условия задачи известны следующие значения:
P(A) = 0.07 - вероятность того, что самолет находится в зоне действия радара,
P(B|A) = 0.98 - вероятность того, что радар показывает наличие самолета, если самолет действительно находится в зоне действия радара.

Чтобы вычислить P(B), используем формулу полной вероятности:
P(B) = P(B|A) * P(A) + P(B|¬A) * P(¬A),

где
P(B|¬A) - вероятность того, что радар показывает наличие самолета, если самолета нет в зоне действия радара,
P(¬A) - вероятность того, что самолета нет в зоне действия радара.

Из условия задачи известно, что P(B|¬A) = 0.05 и P(¬A) = 1 - P(A) = 1 - 0.07 = 0.93.

Теперь можем вычислить P(B):

P(B) = P(B|A) * P(A) + P(B|¬A) * P(¬A)
= 0.98 * 0.07 + 0.05 * 0.93
= 0.0686 + 0.0465
= 0.1151.

Теперь можем вычислить P(A|B) с помощью формулы Байеса:

P(A|B) = (P(B|A) * P(A)) / P(B)
= (0.98 * 0.07) / 0.1151
= 0.0686 / 0.1151
≈ 0.5961.

Таким образом, вероятность того, что самолет находится в зоне действия радара, учитывая, что радар показывает наличие самолета, составляет примерно 0.5961 или около 59.61%.

```python
items_prob = (0.07, 1 - 0.07)
probs = (0.98, 0.05)
bayes_theorem(items_prob, probs, 1, 3, proportion=True)
```
